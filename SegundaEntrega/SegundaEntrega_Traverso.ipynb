# **Ingresos Anuales de los ciudadanos de Estados Unidos**

La intención de este análisis fue saber como es la distribución de los ingresos para los distintos tipos de trabajo y edades entre la población de los nativos y extranjeros en Estados Unidos. El dataset con el que contamos (https://www.kaggle.com/datasets/amirhosseinmirzaie/americancitizenincome) para realizar el análisis no contiene en sus registros datos de una única persona sino que cada registro es una agrupación por edad, tipo de trabajo, educación, estado civil, ocupación, raza, sexo, país origen y tipo de ingreso, el que se clasifica como menor o igual a 50 mil (<=50K) y mayor a 50 mil (>50K), contamos con una campo que nos indica la cantidad de veces que se repite la combinación de datos (fnlwgt).


El cliente tiene unas preguntas especificas a las que le gustaría obtener respuestas y deberá proporcionar visualizaciones para comprender:

* ¿Cual es el porcentaje de horas trabajadas por sexo?
* ¿Como es la distribución de horas según el tipo de ingreso ('<=50K' o '>50K') que tiene cada género?
* ¿Cual es el promedio de horas trabajadas por raza?
* ¿Como es la distribución de personas registradas por razas separada por sexos?
* ¿Como se la distribución de personas de las distintas razas y generos pero mostrando la edad?
* ¿Cual es la distribución de horas trabajadas según la edad de las personas?
* ¿Como son los ingresos según el nivel de educación?
* sobre el nivel de educación ¿Que tipo de tareas realizan los de mayor ingreso? ¿y que ocupaciones tiene?
* sobre los niveles de educación ¿Como es la distribución por edades?
* ¿Como es la distribución de ocupaciones para los nacidos en Estados Unidos y los extranjeros?
* ¿Como es la distribución de horas trabajadas por semana?

Primero analice como es la distribución por sexo y razas, lo que pude ver es que hay una tímida inclución de las mujeres en las actividades laborales casi del 30% de las horas registradas totales, si vemos los mayores ingresos solo hay un 13% de participación del sexo femenino en cambio es del 35% para las ocupaciones de menor ingreso, con esto vemos que si bien hay participación de la mujer esta apertura es mayor para las actividades laborales de menor ingresos.

Si vemos la raza a la que perteneces las personas del dataset y el promedio de horas trabajadas por semana, este es muy similar a todas con valores que estan entre las 42 y 38 horas.

Ahora si vemos la cantidad de personas de cada raza que registran horas la raza blanca es la que mas registra tanto en hombres como mujeres individulamente con respecto a las demás razas. En todas las razas se muestra que el sexo masculino es el que mas hora registras, algo que se deduce de lo que vimos anteriormente. Las razas que siguen en cantidad son afro, asiatica-isleños, americanos-indios-esquimal y otros.

Al ver el boxplot de edad de las personas por raza y sexo
- en las razas White, Asian-Pac-Islander y Others la mediana, y los cuatiles 1 y 3 son menores en mujeres que en hombres, esto por trabajar menos horas por semana en esos grupos.
- todas las razas y sexo tienen el mínimo inlier de 18 años excepto en los hombres de Asian-Pac-Islander que es de 19 aprox.
- en Others vemos que es el grupo que menos horas trabajadas por semana tiene por los máximos de inliers esta en las 60 horas aproximadamente para ambos sexos con algunos muy pocos casos de outliers superior.
- White en ambos géneros es el que mas horas trabajadas tiene con un cuartil 3 mayor y el máximo inlier superior a las demás razas y con varios casos de outliers superiores.

Al ver el histograma de las edades relevada la mayor cantidad de horas regsitradas es aportada por las edades 18 y aproximadamente 54 años.
Y es muy evidente la diferencia que hay entre las 40 horas trabajadas por semana con respecto a las demas cantidades, teniendo outliers de 80 y casi 100 horas.

Cuando empezamos a analizan los ingresos por nivel de estudio alcanzado y la ocupación o tipo de trabajo vemos que para ingresos menores a 50 mil dolares la mayoria de las personas han terminado la escuela secundaria, un bachillerato o egresados universitarios. Para los ingresos mayores a 50 mil se ve que los niveles de educación son similares y se agregan los que realizaron algún master pero la cantidad de personas que termino dicho nivel de educación es muy menor a los que lo terminaron para los ingresos menores.

Al revisar el tipo de trabajo realizado por los individuos de mayores ingresos para los distintos tipos de nivel educacional, la actividad privada es la que se destaca en todos los niveles con mayor o menor cantidad. Las ocupaciones "exec-managerial", "prof-speciality", "sales" y "craft-repair" son los que mas ocurrencia tienen y se relacionan a los niveles de educación "bachelors", "HD-grad", "master", "prof-school" y "some-colleage", con lo que se puede apreciar de que por haber alcanzado un nivel mas alto de estudio ayuda a que se tenga un mayor ingreso.

Para terminar analice la distribución de ocupación y nivel de educación pero separando los nacidos en Estados Unidos y los extranjeros, de estos se ve que la mas alta distribución es "HS-grad", "some-college" y "bachelors" y sobre las ocupaciones "craft-repair", "other-services", "adm-clerical", "prof-specialty" y "exec-managerial". Mientras que para los nacidos en Estados Unidos  los niveles de estudio "HS-grad", "some-college" y "bachelors" es donde se ve que hay mayor cantidad de personas con ingresos mayores y las ocupaciones que sobresalen son "prof-specialty", "exec-manageria", "craft-repair", "adm-clerical" y "sales".

# 2. Distribución de ingresos por sexo y razas

## 2.1 Distribución de horas por sexo
se agruparon los registros por la columna 'sex' y se utiliza la sumatoria de la columa 'horasTrabajadasTotales' porque cada registro es una agrupación.

```python
# @title
figHorasGenero = px.pie(dfAmericanIncomeGraficos,
                        values='horasTrabajadasTotales',
                        names='sex',
                        title='Porcentajes del total de horas por sexo',
                        color='sex',
                        width=1000,
                        color_discrete_map={'Male':'cyan',
                                 'Female':'fuchsia'})
figHorasGenero.show()
```

Podemos ver que falta inclución del sexo femenino en el reparto de horas trabajadas ya que no llega al 30%.

## 2.2 Distribución de horas por sexo para ingresos '>50K'
filtramos el dataset y solo nos quedamos con los registros que contengan el valor '>50K' en la columna 'income' y luego creamos un nuevo dataset donde a lo anterior lo agrupamos por la columna 'sex'

```python
# @title
dfMayores50 = dfAmericanIncomeGraficos.query("income == '>50K'")
dfGroupMayores50Sex = dfMayores50.groupby(['sex'])
dfGroupMayores50SexSum = dfGroupMayores50Sex.sum(numeric_only=True)

figGeneroMayores50 = px.pie(dfGroupMayores50SexSum,
                        values='horasTrabajadasTotales',
                        names=dfGroupMayores50SexSum.index,
                        title='Porcentaje Horas por sexo para ingresos >50K anuales',
                        color=dfGroupMayores50SexSum.index,
                        width=1000,
                        color_discrete_map={'Male':'cyan',
                                 'Female':'fuchsia'})
figGeneroMayores50.show()
```

Lo que podemos ver es que el sexo masculino es el que tiene mas acceso a los trabajos con mayores ingresos y es mas marcada la falta de inclución de la mujer.

## 2.3 Distribución de horas por sexo para ingresos '<=50K'
Como en el gráfico anterior se filtro el dataset original y solo se utilizaron los registros con 'income' '<=50K' agrupandolos luego por el campo 'sex'

```python
# @title
dfMenores50 = dfAmericanIncomeGraficos.query("income == '<=50K'")
dfGroupManores50Sex = dfMenores50.groupby(['sex'])
dfGroupManores50SexSum = dfGroupManores50Sex.sum(numeric_only=True)

figGeneroMenores50 = px.pie(dfGroupManores50SexSum,
                        values='horasTrabajadasTotales',
                        names=dfGroupManores50SexSum.index,
                        title='Porcentaje Horas por sexo para ingresos <=50K anuales',
                        color=dfGroupManores50SexSum.index,
                        width=1000,
                        color_discrete_map={'Male':'cyan',
                                 'Female':'fuchsia'})
figGeneroMenores50.show()
```

Como era de esperar por lo visto en el los dos diagramas de tortas anteriores las mujeres tienen mas acceso a trabajos menos remunerados. No hay gran disperción ya que apenas supera el 35% en ingresos '<=50K', en la distribución general teniamos 29,3% y 13,4% para los mayores ingresos.

## 2.4 Promedio de horas trabajadas por raza
En esta pregunta lo que queremos ver es la cantidad de horas que trabaja cada raza, lo que hacemos es agrupan los registros por el campo 'race' y luego crear un dataset nuevo sumariznado los campos numéricos, el que nos interesa es 'horasTrabajadasTotales' (Cantidad de horas total) al cual lo dividiremos por 'fnlwgt' (cantidad total de los agrupados).

```python
# @title
dfGroupRace = dfAmericanIncomeGraficos.groupby(['race'])
dfGroupRaceSum = dfGroupRace.sum(numeric_only=True)

figGroupRace = px.bar(dfGroupRaceSum,
                      x=dfGroupRaceSum.index,
                      y=(dfGroupRaceSum['horasTrabajadasTotales'] / dfGroupRaceSum['fnlwgt']),
                      color=dfGroupRaceSum.index,
                      width=1000,
                      title='Horas promedio trabajadas por semana por razas',
                      labels={'race': 'Razas', 'y': 'Horas'})
figGroupRace.show()
```

Podemos ver que el promedio de horas por raza es muy similar para todas, al rededor de 40 horas.

## 2.5 Cantidad de horas por razas separada por sexos
creamos un dataset con la cantidad de personas por registro, 'sex' y 'race' y lo agrupamos por 'sex' y 'race' para graficar como es la distribución de genero por raza en el total de las personas registradas en el dataset original.

```python
# @title
dfRazasXSexos = dfAmericanIncomeGraficos[['fnlwgt', 'sex', 'race']].groupby(by=['sex', 'race']).sum(['fnlwgt']).reset_index()

figRazasXSexos = px.bar(dfRazasXSexos,
                        x='race',
                        y='fnlwgt',
                        color='sex',
                        barmode='group',
                        title='Distribución de personas por raza y sexo',
                        width=1200,
                        labels={'race': 'Razas', 'fnlwgt': 'Cant. Personas'},
                        color_discrete_map={'Male':'cyan',
                                 'Female':'fuchsia'})
figRazasXSexos.show()
```

podemos ver que para todas las razas es el hombre el que tiene mas acceso al trabajo registrado aunque no la diferencia con la mujer no es demasiado significativa, siendo la raza Black la que tiene la menor de las diferencias.

# 4. Consumir API

## 4.1 Request al API

```python
# @title
headers = {}
countriesURL  = "https://restcountries.com/v3.1/all?fields=name,continents,independent,unMember,region,languages,area,population,maps"

content = req.get(countriesURL, headers=headers)
result = json.loads(content.content)

independentArr = []
unMemberArr = []
regionArr = []
populationArr = []


for fila in dfIncomesApi.index:

  paisEncontrado = False
  independent = ''
  unMember = ''
  region = ''
  population = ''

  for pais in result:
    reqPais = pais.get('name').get('common')

    if (reqPais == dfIncomesApi['native.country'][fila]) or (reqPais == 'United States' and dfIncomesApi['native.country'][fila] == 'United-States') or (reqPais == 'Trinidad and Tobago' and dfIncomesApi['native.country'][fila] == 'Trinadad&Tobago') or (reqPais == 'Puerto Rico' and dfIncomesApi['native.country'][fila] == 'Puerto-Rico') or (reqPais == 'El Salvador' and dfIncomesApi['native.country'][fila] == 'El-Salvador') or (reqPais == 'Dominican Republic' and dfIncomesApi['native.country'][fila] == 'Dominican-Republic') or (reqPais == 'Hong Kong' and dfIncomesApi['native.country'][fila] == 'Hong'):
      #print(reqPais)
      independent = pais.get('independent')
      unMember = pais.get('unMember')
      region = pais.get('region')
      population = pais.get('population')
      paisEncontrado = True
    elif paisEncontrado == False:
      independent = 'Desconocido'
      unMember = 'Desconocido'
      region = 'Desconocido'
      population = 'Desconocido'

  independentArr.append(independent)
  unMemberArr.append(unMember)
  regionArr.append(region)
  populationArr.append(population)


dfIncomesApi['independent'] = independentArr
dfIncomesApi['unMember'] = unMemberArr
dfIncomesApi['region'] = regionArr
dfIncomesApi['population'] = populationArr

#dfIncomesApi.head(20)
```
